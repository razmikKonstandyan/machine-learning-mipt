```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

Проделаем все на цветочках ириса, при этом будем брать соответственно 2 признака и 3 класса.

```python
iris = datasets.load_iris()
X = iris.data[:, :2] 
Y = iris.target
```

```python
logreg = linear_model.LogisticRegression(solver='sag')
```

```python
logreg.fit(X, Y)

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + 0.5

h = 0.01
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(14, 8))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length', fontsize=14)
plt.ylabel('Sepal width', fontsize=14)

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
```
