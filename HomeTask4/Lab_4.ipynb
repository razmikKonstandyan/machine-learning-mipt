## Task 1

Введем латентные переменные так, чтобы распределение принадлежало экспоненциальному классу: введем переменные
$ z_{dn} \in \{0, 1\}^T$, где $\sum_t z_{dnt} = 1$.
Тогда функция правдоподобия запишется в виде:
$$p(W, Z|\Theta, \Phi) = \prod_{d=1}^{D} \prod_{n=1}^{N_d} \prod_{t=1}^{T} (\theta_{dz_{dnt}} \phi_{z_{dnt} w_{dn}})^{z_{dnt}} $$

Прологарифмируем функцию правдоподобия:
$$ln\ p(W, Z|\Theta, \Phi) = $$
$$ \sum_{d=1}^{D} \sum_{n=1}^{N_d} \sum_{t=1}^{T} z_{dnt}\cdot \ln(\theta_{dz_{dnt}} \phi_{z_{dnt} w_{dn}}) = 
\sum_{d=1}^{D} \sum_{n=1}^{N_d} \sum_{t=1}^{T} z_{dnt}\cdot \ln(\theta_{dz_{t}} \phi_{z_{t} w_{dn}}) = 
\sum_{d=1}^{D} \sum_{n=1}^{N_d} \sum_{t=1}^{T} (z_{dnt}\cdot \ln\theta_{dz_{t}} + z_{dnt}\cdot \ln \phi_{z_{t} w_{dn}})$$

Получили, что $\ln\ p(W, Z|\Theta, \Phi)$ имеет вид $\sum F(W, Z) \cdot G(\Theta, \Phi)$, что значит, что полученное распределение принадлежит экспоненциальному классу.
