```python
import matplotlib.pyplot as plt
import warnings
import pandas as pd
import numpy as np

from collections import OrderedDict
from sklearn import lda, linear_model, svm, datasets, cross_validation
from sklearn.metrics import mean_squared_error

%matplotlib inline
warnings.filterwarnings('ignore')
```

# Подготовим данные.

```python
X = pd.read_csv('data.txt', header=None, sep=' ')
y = pd.read_csv('target.txt', header=None)
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.25, random_state=42)
```

Посчитаем MSE на тесте для линейной регрессии, а так же детерминант матрицы, которую приходится обращать во время обучения.

```python
error = mean_squared_error(y_test, linear_model.LinearRegression().fit(X_train, y_train).predict(X_test))
determinant = np.linalg.det(np.matrix(X_train) * np.matrix(X_train).getT())
print('error:', error) 
print('determinant:', determinant) 
```

Детерминант получился равным нулю, следовательно у нас есть линейно зависимые признаки, а результат получается не надежным.

## Пункт 5

```python
def plot_MSE_alpha(model, alphas, X_train, y_train, X_test, y_test, title):
    result = {}
    for alpha in alphas:
        model.set_params(alpha=alpha).fit(X_train, y_train)
        result[alpha] = mean_squared_error(y_test, model.predict(X_test))
    plt.figure(figsize=(14, 8))
    plt.title(title)
    plt.xlabel('alpha', fontsize=10)
    plt.ylabel('MSE', fontsize=10)
    result = OrderedDict(sorted(result.items()))
    plt.step(result.keys(), result.values(), 'bo')
    return plt
```

```python
plot_MSE_alpha(linear_model.Ridge(), np.arange(0.05, 2, 0.05), X_train, y_train, X_test, y_test, 'RIDGE').show()
```

```python
plot_MSE_alpha(linear_model.Lasso(), np.arange(0.05, 2, 0.05), X_train, y_train, X_test, y_test, 'LASSO').show()
```

```python
plot_MSE_alpha(linear_model.ElasticNet(), np.logspace(-1.5, 1, 100), X_train, y_train, X_test, y_test, 'Elastic Net').show() 
```

Вывод: все три метода выдают разное поведение MSE на изменении параметра alpha. При этом Ridge регрессия показывает логарифмическое увеличение MSE при увеличении alpha, в LASSO линейно убывает, а в Elastic Net сначала убывает и затем возрастает. При этом видно, что самая низкая ошибка достагается на Lasso при увеличении alpha.

## Пункт 6

Для Ridge-регрессии и LASSO-регрессии построим графики траекторий путей весов признаков так, как показано в ноутбуке с семинара 11. Функцию возьмем оттуда же.

```python
def draw_track(model, X, y, log_min, log_max, num, title='', figsize=(12, 6)):
    alphas = np.logspace(log_min, log_max, num)   
    coefs = []
    for a in alphas:
        if 'l1_ratio' in model.get_params():
            l1_ratio = model.get_params()['l1_ratio']
            model.set_params(alpha=0.5 * a * (1 - l1_ratio), l1_ratio=a * l1_ratio)
        else:
            model.set_params(alpha=a)
        model.fit(X, y)
        coefs.append(model.coef_)
    
    plt.figure(figsize=figsize)
    ax = plt.gca()
    ax.hlines(0, 10 ** log_min, 10 ** log_max, linewidth=15, alpha=0.15)
    if title == 'Ridge': coefs = np.squeeze(coefs, axis=(1,2))
    ax.plot(alphas, coefs, linewidth=2)
    ax.set_xscale('log')
    ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
    plt.xlabel('alpha')
    plt.ylabel('weights')
    plt.title(title)
    plt.axis('tight')
    plt.show()
```

```python
draw_track(linear_model.Ridge(fit_intercept=False), X_train, y_train, -1.5, 5, 200, title='Ridge')
draw_track(linear_model.Lasso(fit_intercept=False), X_train, y_train, -2, 2.5, 200, title='Lasso')
```

Вывод: на основе графиков траекторий весов в зависимости от alpha можно сделать вывод, относительно того как алгоритм учитывает данный признак при регрессии. По графикам видно, что в Ridge регрессия, в отличие от Lasso, происходит резкий взрыв весов признаков одновременно. Таким образом Ridge регрессия учитывает практически все признаки, при этом отличия в размерах весов почти нет. По графику видно, что лишь у одного признака вес получился очень большой, а остальные алгоритм прижимает к интервалу около нуля. Если же говорить о Lasso, то тут картина немного другая. А именно, у нас происходит поэтапный отбор признаков. При уменьшении alpha она поэтапно вклучает по одному признаку и наделяет их весами, которые в процессе уменьшения alpha так же изменяются. Во-первых, следует, отметить начилие одного признака, вес которого преобладает над остальными. Во-вторых, в отличие от Ridge регрессии, видно, что тут интервал, в котором разбросаны веса, шире. 

## Пункт 7

Какую функцию потерь используют в SVM регрессии.
Дано: Обучающая выборка $X=\{(x_i,y_i)\}_{i=1}^{\ell}$, где $x_i$-признаковое описание i-го объекта, $y_i$- характеристика, приписываемая объекту.
Функция потерь имеет вид $a(x_i)=\mid (w,x_i)-w_0 \mid_\epsilon$ для каждого вектора $(x_i,y_i)$, где $\mid z \mid_\epsilon = max(0,\mid z \mid-\epsilon)$.

http://scikit-learn.org/stable/modules/svm.html#svm-implementation-details - детали реализации, в том числе функции потерь в sklearn.

```python
def plot_MSE(model, C, c, epsilons, epsilon, X_train, y_train, X_test, y_test, mode):
    res_train, res_test = {}, {}
    for eps in epsilons:
        model.set_params(C=c, epsilon=eps).fit(X_train, y_train)
        if mode == 'MSE':
            res_train[eps] = mean_squared_error(y_train, model.predict(X_train))
            res_test[eps] = mean_squared_error(y_test, model.predict(X_test))
        elif mode == 'LossFunction':
             res_train[eps] = max(np.array([abs(np.matrix(X_train[i])*np.matrix(model.coef_.reshape(25, 1))-
                                  model.intercept_-y_train[i])-eps for i in range(len(X_train))]).mean(), 0)
             res_test[eps] = max(np.array([abs(np.matrix(X_test[i])*np.matrix(model.coef_.reshape(25, 1))-
                                  model.intercept_-y_test[i])-eps for i in range(len(X_test))]).mean(), 0)
    res_train = OrderedDict(sorted(res_train.items()))
    res_test = OrderedDict(sorted(res_test.items()))
    f, axarr = plt.subplots(2, 2, figsize=(18, 10))
    axarr[0, 0].step(res_train.keys(), res_train.values(), 'bo')
    axarr[0, 0].set_title('Fixed C, train data')   
    axarr[0, 1].step(res_test.keys(), res_test.values(), 'bo')
    axarr[0, 1].set_title('Fixed C, test data')
    res_train, res_test = {}, {}
    for c in C:
        model.set_params(C=c, epsilon=epsilon).fit(X_train, y_train)
        if mode == 'MSE':
            res_train[c] = mean_squared_error(y_train, model.predict(X_train))
            res_test[c] = mean_squared_error(y_test, model.predict(X_test))
        elif mode == 'LossFunction':
            res_train[c] = max(np.array([abs(np.matrix(X_train[i])*np.matrix(model.coef_.reshape(25, 1))-
                                  model.intercept_-y_train[i])-epsilon for i in range(len(X_train))]).mean(), 0)
            res_test[c] = max(np.array([abs(np.matrix(X_test[i])*np.matrix(model.coef_.reshape(25, 1))-
                                  model.intercept_-y_test[i])-epsilon for i in range(len(X_test))]).mean(), 0)
    res_train = OrderedDict(sorted(res_train.items()))
    res_test = OrderedDict(sorted(res_test.items()))
    axarr[1, 0].step(res_train.keys(), res_train.values(), 'bo')
    axarr[1, 0].set_title('Fixed epsilon, train data')   
    axarr[1, 1].step(res_test.keys(), res_test.values(), 'bo')
    axarr[1, 1].set_title('Fixed epsilon, test data') 
    plt.show()
```

```python
plot_MSE(svm.SVR(), np.arange(0.2, 2, 0.2), 1.0, np.arange(0.05, 0.9, 0.05), 0.1, X_train, y_train, X_test, y_test,
        'MSE')
```

```python
plot_MSE(svm.SVR(kernel='linear'), np.arange(0.2, 2, 0.2), 1.0, np.arange(0.05, 0.9, 0.05), 0.1, 
         np.array(X_train), np.array(y_train), np.array(X_test), np.array(y_test),
        'LossFunction')
```

Из вышестоящих графиков оптимально брать побольше С и побольше epsilon. На MSE видно, что epsilon почти не влияет. А вот при увеличении C MSE уменьшается. Противоположная картина для функции потерь.
