```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.grid_search import GridSearchCV
%matplotlib inline
```

# Регрессия
В качестве функции возьмем расстояние от точки до центра координат.

```python
X_train = scipy.stats.multivariate_normal.rvs(size=200, mean=[0, 0], cov=[[3, 0.1], [0.1, 2.8]])
y_train = (X_train[:, 0] ** 2 + X_train[:, 1] ** 2) ** 0.5

plt.figure(figsize=(10, 8))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='summer', s=70)
plt.show()
```

Обучение модели

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

Посмотрим на значение в нуле

```python
model.predict([[0, 0]])
```

Среднеквадратичная ошибка

```python
mean_squared_error(y_train, model.predict(X_train))
```

Визуализация

```python
def generate_grid(train_sample, border=1, step=0.05):
    return np.meshgrid(np.arange(min(train_sample[:, 0]) - border, max(train_sample[:, 1]) + border, step), 
                       np.arange(min(train_sample[:, 1]) - border, max(train_sample[:, 1]) + border, step))
```

```python
def create_picture(X_train, y_train, model, border=1, step=0.05, cmap='summer', 
                   alpha=1, create_new_figure=True, figsize=(10, 5)):
    
    # Создание сетки
    grid = generate_grid(X_train, border, step)
    grid_ravel = np.c_[grid[0].ravel(), grid[1].ravel(0)]  # Выворачивание сетки
    
    # Предсказание значений для сетки
    grid_predicted_ravel = model.predict(grid_ravel)
    grid_predicted = grid_predicted_ravel.reshape(grid[0].shape)  # Подгоняем размер
    
    
    # Построение фигуры
    if create_new_figure:
        plt.figure(figsize=figsize)
        
    plt.pcolormesh(grid[0], grid[1], grid_predicted, cmap=cmap)
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, alpha=alpha, cmap=cmap)
    plt.xlim((min(grid_ravel[:, 0]), max(grid_ravel[:, 0])))
    plt.ylim((min(grid_ravel[:, 1]), max(grid_ravel[:, 1])))
    plt.title('max_depth = ' + str(model.get_params()['max_depth']) + \
              ', min_samples_leaf = ' + str(model.get_params()['min_samples_leaf']))
    plt.xlabel('$x_0$')
    plt.ylabel('$x_1$')
    
    if create_new_figure:
        plt.show()
```

Получаем сильно изрезанную картинку.
Можель выдает кусочно-постоянную функцию --- дерево все-таки.

```python
create_picture(X_train, y_train, RandomForestRegressor().fit(X_train, y_train))
```

# Классификация

```python
from sklearn.datasets import make_blobs
from sklearn.ensemble import RandomForestClassifier
```

Сгенерируем три класса

```python
X_train, y_train = make_blobs(n_samples=300, cluster_std=3)

plt.figure(figsize=(15, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, alpha=0.5, cmap='spring', s=70)
```

Обучим простое дерево и посмотрим на результат классификации

```python
model = RandomForestClassifier(min_samples_leaf=5, max_features=2).fit(X_train, y_train)
create_picture(X_train, y_train, model, cmap='spring')
```
